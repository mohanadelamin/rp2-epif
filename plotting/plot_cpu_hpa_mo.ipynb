```python


def get_time_in_s_stat(x, start_point):
    x = x.split("T")[-1]
    x = datetime.strptime(x, '%H:%M:%S.%f')
    min_two_hours = x - timedelta(hours=2, minutes=0)
    x = datetime.strptime(min_two_hours.strftime("%H:%M:%S.%f"), '%H:%M:%S.%f')
    a_timedelta = x - datetime(1900, 1, 1)
    seconds = a_timedelta.total_seconds()

    return seconds - start_point



def get_stat(data_directory, start_point):
    df_stat = pd.read_csv(f"{data_directory}/hpa_stats.txt", sep=",", names=["Time", "currentReplicas","cpuPercentage"])
    df_stat = df_stat.dropna()
    df_stat = df_stat.reset_index()
    df_stat["Time_in_s"] = df_stat["Time"].apply(lambda x: get_time_in_s_stat(str(x), start_point))
    df_stat["Time_in_s"] =  df_stat["Time_in_s"] 
    return df_stat
 

def plot_stats(df_stat):
    plt.title(f"Number of pods")
        
    # Plot the cpu at a gradient of 1 precent, get microseconds of CPU per s (milicore)
    plt.plot(df_stat["Time_in_s"], df_stat["currentReplicas"], label="Number of pods")
    plt.xlabel("Second")
    plt.ylabel("Number of pods")
    plt.legend(loc="upper left")
    plt.show()
```

```python
def get_time_in_vms_stat(x, start_point):
    x = x.split("T")[-1][:-7]
    x = datetime.strptime(x, '%H:%M:%S.%f')
#     min_two_hours = x - timedelta(hours=2, minutes=0)
#     x = datetime.strptime(min_two_hours.strftime("%H:%M:%S.%f"), '%H:%M:%S.%f')
    a_timedelta = x - datetime(1900, 1, 1)
    seconds = a_timedelta.total_seconds()

    return seconds - start_point



def get_vms_stat(data_directory, start_point):
    vm_stat = pd.read_csv(f"{data_directory}/vms_stats.csv", skiprows=1, sep=",", names=["timestamp","node","cpu_seconds","cpu_percentage","mem_k","mem_percentage"])
    vm_stat = vm_stat.dropna()
    vm_stat = vm_stat.reset_index()
    vm_stat["Time_in_s"] = vm_stat["timestamp"].apply(lambda x: get_time_in_vms_stat(str(x), start_point))
    vm_stat["Time_in_s"] = vm_stat["Time_in_s"] 
    return vm_stat
 

def plot_vms_stats(vm_stat):
    plt.title(f"K8S Worker nodes CPU")
        
    # Plot the cpu at a gradient of 1 precent, get microseconds of CPU per s (milicore)
    worker_set = set(vm_stat["node"])
    
    for worker in worker_set:
        worker_time_list = []
        worker_cpu_list = []
        for w in range(len(vm_stat["node"])):
            if worker == vm_stat["node"][w]:
                worker_time_list.append(vm_stat["Time_in_s"][w])
                worker_cpu_list.append(vm_stat["cpu_seconds"][w])
        plt.plot(worker_time_list, worker_cpu_list, label=worker)
    plt.xlabel("Second")
    plt.ylabel("CPU")
    plt.legend(loc="upper left")
    plt.show

            
#     plt.plot(vm_stat["Time_in_s"], vm_stat["cpu_seconds"])
#     plt.xlabel("Second")
#     plt.ylabel("CPU")
#     plt.show()
```

```python
def get_datetime_response_time(x, start_time):
#     x = datetime.fromtimestamp(x)
#     x = x.strftime('%H:%M:%S.%f')
#     x = datetime.strptime(x, "%H:%M:%S.%f")
    x = x - start_time
    return x
    
def get_response_datetime(df):
    initial_value = df['time'].iloc[0]
#     df['time_datetime'] = df['time'].apply(lambda x: get_datetime_response_time(x, df['time_datetime'][0]))
    df['time_datetime'] = df['time'].apply(lambda x: get_datetime_response_time(x, initial_value))
    return df 

def plot_response_time(data_directory,df_stat):
    plt.title("Response time")
    plot_hline_for_scaling_event(df_stat)
    df = pd.read_csv(f"{data_directory}/response_times", names=["time", "respone_time"])
    #print(df['time'])
    df = get_response_datetime(df)
    df = df.set_index("time_datetime")
    #df = df.resample("1s").mean()
    plt.xticks(rotation=30)
    plt.xlabel("Timestamp")
    plt.ylabel("response_time")
    plt.plot(df.index, df["respone_time"], label=f"Response Time")
    plt.legend(loc="upper right")
    plt.xlim(0,100)
    plt.show()
```
